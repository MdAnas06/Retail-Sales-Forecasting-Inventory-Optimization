```python
# Retail Sales Forecasting & Inventory Optimization
# Objective: Predict demand and optimize inventory decisions
import pandas as pd

df = pd.read_csv("../data/processed/feature_data.csv")
forecast = pd.read_csv("../outputs/forecast_results.csv")
inventory = pd.read_csv("../outputs/inventory_plan.csv")

df.head()
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df['date'], df['sales'])
plt.title("Sales Trend")
plt.show()
plt.figure(figsize=(12,5))
plt.plot(forecast['actual_sales'], label='Actual')
plt.plot(forecast['predicted_sales'], label='Predicted')
plt.legend()
plt.title("Forecast Performance")
plt.show()
plt.figure(figsize=(12,5))
plt.plot(inventory['current_stock'], label='Stock')
plt.title("Inventory Levels")
plt.legend()
plt.show()
reorders = inventory[inventory['reorder_flag'] == True]
reorders.head()

```
